```python
cdf.drop(['Sno', 'Time', 'ConfirmedIndianNational', 'ConfirmedForeignNational'], inplace = True, axis = 1)
```

```python
cdf.head(10)
```

```python
cdf['Date'] = pd.to_datetime(cdf['Date'], format = '%Y-%m-%d')
```

```python
cdf.head()
```

```python
# Active Cases = Confirmed Cases - (Total no. of cured + deaths)

cdf['Active_Cases'] = cdf['Confirmed'] - (cdf['Cured'] + cdf['Deaths'])
cdf.tail()
```

```python
statewise = pd.pivot_table(cdf, values = ['Confirmed', 'Deaths', 'Cured'], index = 'State/UnionTerritory', aggfunc = max) 
```

```python
statewise['Recovery Rate'] = statewise['Cured']*100 / statewise['Confirmed']
```

```python
statewise['Mortality Rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise
```

```python
statewise = statewise.sort_values(by = 'Confirmed', ascending = False)
```

```python
statewise.style.background_gradient(cmap = 'cubehelix')
```

```python
# Top 10 Active Cases

top_10_active_cases = cdf.groupby(by = 'State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by = ['Active_Cases'], ascending = False).reset_index() 
fig = plt.figure(figsize = (16, 9))
plt.title("Top 10 states with most active cases in India", size = 25)
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_Cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = 'yellow')
plt.xlabel("State/UnionTerritory")
plt.ylabel("Active_Cases")
plt.show()
```

```python
# Top states with highest deaths

top_10_death_cases = cdf.groupby('State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = 'Deaths', ascending = False).reset_index()
fig = plt.figure(figsize = (16,5))
plt.title("Top 10 states with highest number of deaths in India", size = 25)
ax = sns.barplot(data = top_10_death_cases.iloc[:12], x = 'State/UnionTerritory', y = 'Deaths', linewidth = 2, edgecolor = 'grey')
plt.xlabel("State/UnionTerritory")
plt.ylabel("Active_Cases")
plt.show()
```

```python
# Growth trend

fig = plt.figure(figsize = (12, 6))
some_states = cdf['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])
ax = sns.lineplot(x = cdf[some_states]['Date'], y = cdf[some_states]['Deaths'], hue = cdf[some_states]['State/UnionTerritory'])
ax.set_title("Top 5 affected States in India", size = 16)
plt.show()
```

```python
# Remove rows where State = India

vaccine = vdf[vdf.State!= 'India']
vaccine
```

```python
vaccine.rename(columns = {'Total Individuals Vaccinated': "Total"}, inplace = True)
```

```python
# Most Vaccinated State

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')     # Series.to_frame() is used to convert a series into a DataFrame
max_vac = max_vac.sort_values('Total', ascending = False)[:5]
max_vac
```

```python
# plot for most vaccinated states

fig = plt.figure(figsize = (10,5))
plt.title('Top 5 Vaccinated States in India', size = 20)
ax = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth = 2, edgecolor = 'grey')
plt.xlabel('States')
plt.ylabel('Vaccinattion')
plt.show()
```

```python
# Least 5 vaccinated states in India

least_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
least_vac = least_vac.sort_values('Total', ascending = True)[:5]
least_vac
```

```python
# plot for least 5 vaccinated states

fig = plt.figure(figsize = (10,5))
plt.title('Least 5 Vaccinated States in India', size = 20)
ax = sns.barplot(data = least_vac.iloc[:10], y = least_vac.Total, x = least_vac.index, linewidth = 2, edgecolor = 'grey')
plt.xlabel('States')
plt.ylabel('Vaccinattion')
plt.show()
```
